```python
import numpy as np
import numpy.ma as ma
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from mpl_toolkits.axes_grid.anchored_artists import AnchoredText
```

```python
from uncertainties import ufloat, unumpy
from ToyMC.utils.Utils import is_logarithmic, get_bin_centers, get_bin_sizes
```

```python
LEVEL = 0
NAME = '14500'
INFILE = NAME+'.pkl'
WEIGHT_FILE = NAME+'_weights.hdf5'
YS_BINNING = np.linspace(0, 1, 11)
#XS_CENTERS = np.array([0.015, 0.045, 0.125, 0.175, 0.275, 0.35, 0.55, 0.65])
XS_BINNING = np.logspace(-1.8, -0.10, 6)
E_BINNING = np.logspace(-1, 3, 4)
```

```python
infile = '/data/icecube/data/mlarson/level7_24Nov2015/GENIE_parms/level'+str(LEVEL)+'/'+INFILE
inweights = '/data/icecube/data/mlarson/level7_24Nov2015/GENIE_parms/level'+str(LEVEL)+'/'+WEIGHT_FILE
input_file = pickle.load(open(infile, 'rb'))
if '16' in NAME:
    weights = np.ones(len(input_file['cc']))
else:
    weights = np.array(h5py.File(inweights, 'r')['weights'][:])
```

```python
print(sorted(input_file.keys()))
print(weights.shape)
```

```python
if not np.sum(weights) == len(input_file['cc']):
    weights *= 3600 * 24 * 365.
    if LEVEL == 0:
        weights /= 3.
    elif LEVEL == 7:
        if '12585' in INFILE:
            weights /= 275.
        elif '14585' in INFILE:
            weights /= 245.
else:
    print(weights)
```

```python
def format_axis(ax, x_label=r'', title=r'', lhs=False, ylim=None):
    ax.set_title(title)

    ax.set_xlim(np.min(YS_BINNING), np.max(YS_BINNING))
    for xmaj in ax.xaxis.get_majorticklocs():
        ax.axvline(x=xmaj, ls=':', color='gray', alpha=0.4, linewidth=1)
    if x_label == r'':
        ax.get_xaxis().set_ticks([])
    else:
        ax.set_xlabel(x_label, fontsize=14)
    if is_logarithmic(YS_BINNING):
        ax.set_xscale('log')

    ax.tick_params(axis='y', labelsize=10)
    if not ylim is None:
        ax.set_ylim(ylim)
    #ax.set_ylim(1, 7000)
    #ax.set_yscale('log')
    for ymaj in ax.yaxis.get_majorticklocs():
        ax.axhline(y=ymaj, ls=':', color='gray', alpha=0.4, linewidth=1)
    if not lhs:
        ax.get_yaxis().set_ticks([])
        xticks = ax.xaxis.get_major_ticks()
        try:
            xticks[0].set_visible(False)
        except:
            pass
    else:
        yticks = ax.yaxis.get_major_ticks()
        yticks[0].set_visible(False)

def plot_histo(ax, array, weights, xsec_weights, divide_factors, colour):
    #print('xsec_weights', xsec_weights)
    hist, edges = np.histogram(array, bins=YS_BINNING, weights=weights)
    hist_2, edges = np.histogram(array, bins=YS_BINNING, weights=weights**2)
    hist_w, edges = np.histogram(array, bins=YS_BINNING, weights=weights*xsec_weights)
    hist_w2, edges = np.histogram(array, bins=YS_BINNING, weights=(weights*xsec_weights)**2)

    u_hist = unumpy.uarray(hist, np.sqrt(hist_2))
    u_hist_w = unumpy.uarray(hist_w, np.sqrt(hist_w2))
    #dsigma = u_hist_w / u_hist
    dsigma, missing = [], []
    for idx in xrange(len(u_hist)):
        if u_hist[idx].n < 1E-6:
            dsigma.append(ufloat(0, 0))
            missing.append(1)
            continue
        missing.append(0)
        dsigma.append(u_hist_w[idx] / u_hist[idx])
    dsigma, missing = np.array(dsigma), np.array(missing)
    #print('dsigma', unumpy.nominal_values(dsigma))

    b_sizes = get_bin_sizes(edges)
    #print('b_sizes', b_sizes)
    #print('divide_factors', divide_factors)
    #print('b_sizes * divide_factors', b_sizes*divide_factors)
    dsigma_dxdy = (dsigma / 1E-38) / (b_sizes.astype(float) * divide_factors)
    #print('dsigma_dxdy', unumpy.nominal_values(dsigma_dxdy))

    dsigma_dxdy_0 = np.concatenate(([dsigma_dxdy[0]], dsigma_dxdy))
    missing_0 = np.concatenate(([missing[0]], missing))
    ax.step(
        YS_BINNING, ma.masked_array(unumpy.nominal_values(dsigma_dxdy_0), mask=missing_0),
        alpha=0.5, color=colour, drawstyle='steps-mid', linewidth=2, linestyle=':'
    )

    ax.errorbar(
        YS_BINNING, ma.masked_array(unumpy.nominal_values(dsigma_dxdy_0), mask=missing_0),
        xerr=0, yerr=ma.masked_array(unumpy.std_devs(dsigma_dxdy_0), mask=missing_0),
        capsize=0.7, alpha=1, color=colour, linestyle='none', markersize=0.5, linewidth=1
    )
```

```python
pdg_array = input_file['neu']
xsec_array = input_file['xsec'] * 1E-38
energy_array = input_file['energy']
ys_array = input_file['y']
xs_array = input_file['x']
dis_array = input_file['dis']
nu_mask = pdg_array > 0
nubar_mask = pdg_array < 0
print(xsec_array)
```

```python
e_bin_centers = get_bin_centers(E_BINNING)
x_bin_centers = get_bin_centers(XS_BINNING)
x_bin_sizes = get_bin_sizes(XS_BINNING)
for e_idx, e_bin in enumerate(E_BINNING[:-1]):
    #e_cont = True
    #for x in E_PLOT:
    #    if x >= e_bin and x < E_BINNING[e_idx + 1]:
    #        e_cont = False
    #if e_cont:
    #    continue
    print(e_idx, e_bin_centers[e_idx])
    energy_mask = (energy_array[dis_array] > e_bin) & (energy_array[dis_array] < E_BINNING[e_idx + 1])
    xsec_weight = xsec_array[dis_array][energy_mask]
    ys_array_e = ys_array[dis_array][energy_mask]
    xs_array_e = xs_array[dis_array][energy_mask]
    weights_e = weights[dis_array][energy_mask]
    nu_mask_e = nu_mask[dis_array][energy_mask]
    nubar_mask_e = nubar_mask[dis_array][energy_mask]

    fig = plt.figure(figsize=[12, 14])
    if LEVEL == 0:
        fig.suptitle('Level0 (Generator level) '+NAME, y=0.93)
    elif LEVEL == 7:
        fig.suptitle('Level7 (Final level) '+NAME, y=0.93)
    gs = gridspec.GridSpec(len(XS_BINNING)-1, 1)
    gs.update(hspace=0., wspace=0.)
    
    for x_idx, x_bin in enumerate(XS_BINNING[:-1]):
        #if x_idx == 1: break
        xs_mask = (xs_array_e > x_bin) & (xs_array_e < XS_BINNING[x_idx+1])
        ys_array_e_x = ys_array_e[xs_mask]
        xsec_weight_x = xsec_weight[xs_mask]
        weights_e_x = weights_e[xs_mask]
        nu_mask_e_x = nu_mask_e[xs_mask]
        nubar_mask_e_x = nubar_mask_e[xs_mask]
        
        gs0 = gridspec.GridSpecFromSubplotSpec(1, 2, subplot_spec=gs[x_idx],
                                               wspace=0, hspace=0, width_ratios=[1,1])
        ax0 = fig.add_subplot(gs0[0])
        plot_histo(
            ax0, ys_array_e_x[nu_mask_e_x], weights_e_x[nu_mask_e_x],
            xsec_weight_x[nu_mask_e_x],
            divide_factors=e_bin_centers[e_idx]*x_bin_sizes[x_idx], colour='red'
        )
        if x_idx == 0:
            format_axis(ax0, title=r'Neutrino', lhs=True)
        elif x_idx == len(XS_BINNING)-2:
            format_axis(ax0, x_label=r'Y', lhs=True)
        else:
            format_axis(ax0, lhs=True)
            
        ax1 = fig.add_subplot(gs0[1])
        plot_histo(
            ax1, ys_array_e_x[nubar_mask_e_x], weights_e_x[nubar_mask_e_x],
            xsec_weight_x[nubar_mask_e_x],
            divide_factors=e_bin_centers[e_idx]*x_bin_sizes[x_idx], colour='blue'
        )
        if x_idx == 0:
            format_axis(ax1, title=r'Anti-Neutrino')
        elif x_idx == len(XS_BINNING)-2:
            format_axis(ax1, x_label=r'Y')
        else:
            format_axis(ax1)
        
        ylim = (np.min([ax0.get_ylim()[0], ax1.get_ylim()[0]]), np.max([ax0.get_ylim()[1], ax1.get_ylim()[1]]))
        ax0.set_ylim(ylim)
        ax1.set_ylim(ylim)
        
        at = AnchoredText(r'xs = {0:.2f}'.format(x_bin_centers[x_idx]),
                          prop=dict(size=10), frameon=True, loc=1)
        at.patch.set_boxstyle("round,pad=0.,rounding_size=0.2")
        ax1.add_artist(at)
        
        fig.text(0.055, 0.5, r'$\rm{1/E d}^2\sigma/\rm{dx dy } (\times 10^{-38}\rm{cm}^2/\rm{GeV})$',
                 rotation='vertical', va='center', size=20)
        fig.text(0.5, 0.075, r'(E = {0:.2f} GeV)'.format(e_bin_centers[e_idx]), ha='center', size=20)
        
    fig.savefig('./images/NuTeV/level'+str(LEVEL)+'/'+NAME+'_'+'{0:.2f}_.png'.format(e_bin_centers[e_idx]),
                bbox_inches='tight')
```
